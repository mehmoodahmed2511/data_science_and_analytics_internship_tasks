```python
import pandas as pd          # importing pandas library
import matplotlib.pyplot as plt   # importing matplotlib.pyplot for using visualizations 
import seaborn as sns   # importing seaborn for using visualizations
```

```python
df = pd.read_csv ('irisraw.csv')          # loading iris dataset csv file using pandas read_csv function
```

```python
print(df.shape)          # finding rows and columns of iris dataset using shape attribute
```

```python
print(df.columns)          # finding iris dataset columns names using columns attribute
```

```python
print(df.head())          # 5 starting rows of this dataset using head() function
```

```python
# Correlation between Sepal Length and Sepal Width using Matplotlib Scatter Plot

plt.scatter('sepal length', 'sepal width', data = df)
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('Scatter Plot')
```

```python
# Correlation between Petal Length and Petal Width using Seaborn Scatter Plot

sns.scatterplot(data = df, x='petal length', y='petal width', hue='species') 
```

```python
# Correlation between Sepal Length and Petal Length using Matplotlib Scatter Plot
plt.scatter(x='sepal length', y='petal length', data =df)
plt.title('Scatter Plot (correlation between Sepal Length and Petal Length)')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Length')
```

```python
# Sepal Length visual using Seaborn Histogram
sns.histplot(x='sepal length', data=df, kde=True)
```

```python
# Petal Length visual using Matplotlib Histogram
plt.hist(x='petal length', data=df)
plt.title('Histogram')
plt.xlabel('Petal length')
plt.ylabel('Count')
```

```python
# Species categories visual using Seaborn Historgram
sns.histplot(x='species', data=df, kde=True)
```
